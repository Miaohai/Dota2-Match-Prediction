### 2. Hero Pair Win Rate 

#####    Algorithm:
        loop through dict1 which contains all the hero pair
            loop through all the radiant teams which won the game
                if hero pair exists in the team then Win count +=1, Total mathches +=1
            loop through all the dire teams which lose the game
                if hero pair exists in the team then Total mathches +=1

```python
heroes= pd.read_csv('heroes.csv')
import itertools
x=itertools.combinations(heroes['Hero Name'], 2)
dict1={}
for i in x:
    dict1[i]={"Win":0, "Total":0}
    
for i in dict1:
    count=0
    for j in radiant_win['radiant']:
        if (i[0] in j and i[1] in j):
            count+=1
    dict1[i]['Win']=count
    dict1[i]['Total']=count+dict1[i]['Total']
    
    count1=0
    for j in radiant_win['dire']:
        if (i[0] in j and i[1] in j):
            count1+=1
    dict1[i]['Total']=count1+dict1[i]['Total']
```

Retrieve all the matches that dire won the match

```python
dire_win=X[X["radiant_win"]==0]
dire_win
```

#####    Algorithm:
        loop through dict1 which contains all the hero pair
            loop through all the radiant teams which lose the game
                if hero pair exists in the team then Total mathches +=1
            loop through all the dire teams which won the game
                if hero pair exists in the team then Win count +=1 Total mathches +=1

```python
for i in dict1:
    count=0
    for j in dire_win['radiant']:
        if (i[0] in j and i[1] in j):
            count+=1
    dict1[i]['Total']=count+dict1[i]['Total']
    
    count1=0
    for j in dire_win['dire']:
        if (i[0] in j and i[1] in j):
            count1+=1
    dict1[i]['Win']=count1+dict1[i]['Win']
    dict1[i]['Total']=count1+dict1[i]['Total']
```

### 3. Top 10 Hero Pair Win Rate

```python
hero_pair_win_rate=pd.DataFrame.from_dict(dict1)
new=hero_pair_win_rate.transpose()
new.head()
```

```python
def winrate(x):
    new = x.copy()
    ### Your code goes here
    new['winrate'] = new['Win'] / new.Total
    return new
new1=winrate(new)

# Select hero pair that played more than 50 matches
more_than_50_matches = new1[new1['Total']>=50]

more_than_50_matches=more_than_50_matches.sort_values(by=['winrate'],ascending=False,)
top10 = more_than_50_matches.head(10)
top10
```

### 4.  Calculate Synergy for each team and their differences 

#### Adding the Synergy of each team in the dataframe

```python
def radiant(i):
    synergy = 0
    i.sort()
    x=itertools.combinations(i, 2)
    for j in x:
        if (j[0] == 'Unknown' or j[1] == 'Unknown'):
            synergy+=0
        else:
            try:
                synergy+=dict1[j]['Win']/dict1[j]['Total']
            except:
                synergy+=dict1[j[1],j[0]]['Win']/dict1[j[1],j[0]]['Total']
    return synergy
    

X['synergy_radiant']=X['radiant'].apply(radiant)


def dire(i):
    synergy = 0
    i.sort()
    x=itertools.combinations(i, 2)
    for j in x:
        if (j[0] == 'Unknown' or j[1] == 'Unknown'):
            synergy+=0
        else:
            try:
                synergy+=dict1[j]['Win']/dict1[j]['Total']
            except:
                synergy+=dict1[j[1],j[0]]['Win']/dict1[j[1],j[0]]['Total']
    return synergy
    

X['synergy_dire']=X['dire'].apply(dire)
```

```python
X.head()
```

We can then construct a single feature which represents the
difference between hero synergy on the radiant and dire side

```python
def difference(x):
    new_df = x.copy()
    ### Your code goes here
    new_df['df_between_hero_synergy'] = new_df.synergy_radiant - new_df.synergy_dire
    return new_df

new_df=difference(X)
```
