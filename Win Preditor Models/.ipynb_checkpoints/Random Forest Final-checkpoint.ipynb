```python
import pandas as pd
import numpy as np
```

```python
X = pd.read_csv("final_dataset.csv")
matches = pd.read_csv('match.csv')
y = matches['radiant_win'].apply(lambda win: 1 if win else 0)
```

```python
X_counter = X.drop('df_between_hero_synergy', axis=1)
X_counter
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 200, max_features=25, random_state = 7)

X_counter_train, X_counter_test, y_train, y_test = train_test_split(X_counter, y, test_size=0.25, random_state=7)

clf.fit(X_counter_train, y_train)
clf.score(X_counter_test, y_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 200, max_features=27, random_state = 7)

X_counter_train, X_counter_test, y_train, y_test = train_test_split(X_counter, y, test_size=0.25, random_state=7)

clf.fit(X_counter_train, y_train)
clf.score(X_counter_test, y_test)
```

```python
X_synergy = X.drop('diff', axis=1)
X_synergy
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 200, max_features=25, random_state = 7)

X_synergy_train, X_synergy_test, y_train, y_test = train_test_split(X_synergy, y, test_size=0.25, random_state=7)

clf.fit(X_synergy_train, y_train)
clf.score(X_synergy_test, y_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 200, max_features=30, random_state = 7)

X_synergy_train, X_synergy_test, y_train, y_test = train_test_split(X_synergy, y, test_size=0.25, random_state=7)

clf.fit(X_synergy_train, y_train)
clf.score(X_synergy_test, y_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 200, max_features=25, random_state = 7)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=7)

clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 200, max_features=25, random_state = 7)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=7)

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print('accuracy:',accuracy_score(y_test, y_pred))
```

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.metrics import roc_auc_score, roc_curve

# predict probabilities
probs = clf.predict_proba(X_test)
# keep probabilities for the positive outcome only
probs = probs[:, 1]

auc = roc_auc_score(y_test, probs)
print('AUC - Test Set: %.2f%%' % (auc*100))

# calculate roc curve
fpr, tpr, thresholds = roc_curve(y_test, probs)
# plot no skill
plt.plot([0, 1], [0, 1], linestyle='--')
# plot the roc curve for the model
plt.plot(fpr, tpr, marker='.')
# show the plot
plt.show()
```

```python
predictions = clf.predict(X_test)

from sklearn.metrics import precision_recall_fscore_support as score, precision_score, recall_score, f1_score
precision = precision_score(y_test, predictions)
print('Precision: %f' % precision)
# recall: tp / (tp + fn)
recall = recall_score(y_test, predictions)
print('Recall: %f' % recall)
# f1: tp / (tp + fp + fn)
f1 = f1_score(y_test, predictions)
print('F1 score: %f' % f1)
```

```python
from sklearn.metrics import log_loss
accuracy = log_loss(y_test, predictions)
print("Logloss: %.2f" % (accuracy))
```
